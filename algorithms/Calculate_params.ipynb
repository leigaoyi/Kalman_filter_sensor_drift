```python
import numpy as np
import matplotlib.pyplot as plt

# example data
x = np.arange(0.1, 4, 0.5)
y = np.exp(-x)

# example error bar values that vary with x-position
error = 0.1 + 0.2 * x

fig, (ax0, ax1) = plt.subplots(nrows=2, sharex=True)
ax0.errorbar(x, y, yerr=error, fmt='-o')
ax0.set_title('variable, symmetric error')

# error bar values w/ different -/+ errors that
# also vary with the x-position
lower_error = 0.4 * error
upper_error = error
asymmetric_error = [lower_error, upper_error]

ax1.errorbar(x, y, xerr=asymmetric_error, fmt='o')
ax1.set_title('variable, asymmetric error')
ax1.set_yscale('log')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch

# Create a sine-like curve
fs = 1000  # Sampling frequency
T = 1  # Duration in seconds
t = np.linspace(0, T, T * fs, endpoint=False)  # Time vector
f0 = 10  # Frequency of the sine wave
x = np.sin(2 * np.pi * f0 * t)  # Sine-like curve
print(x.shape)
# Compute the power spectral density using Welch's method
f, Pxx = welch(x, fs, nperseg=1024)

# Plot the power spectral density
plt.figure()
plt.semilogy(f, Pxx)
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power Spectral Density [V**2/Hz]')
plt.title('PSD of a Sine-like Curve (Log scale)')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate the sine-like curve
x = np.linspace(0, 2*np.pi, 1000)
y = np.sin(x)

# Calculate the power spectrum density of the curve
fft_y = np.fft.fft(y)
psd = np.abs(fft_y)**2 / len(y)
freqs = np.fft.fftfreq(len(y), x[1]-x[0])
idx = np.argsort(freqs)

# Plot the power spectrum density
plt.plot(freqs[idx], psd[idx])
plt.xlabel('Frequency')
plt.ylabel('Power Spectrum Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
f_mod = 1000   # modulation frequency
f_sig = 100    # signal frequency
A_sig = 1      # signal amplitude
phi = np.pi/2  # signal phase

# Generate signal
t = np.arange(0, 1, 1/10000)
sig = A_sig * np.sin(2 * np.pi * f_sig * t + phi)

# Generate reference signal
ref = np.sin(2 * np.pi * f_mod * t)

# Demodulate signal
I = np.mean(sig * ref)  # in-phase component
Q = np.mean(sig * (ref * 1j))  # quadrature component

# Plot results
fig, axs = plt.subplots(3, 1, figsize=(10, 8))
axs[0].plot(t, sig)
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Amplitude')
axs[0].set_title('Signal')
axs[1].plot(t, ref)
axs[1].set_xlabel('Time (s)')
axs[1].set_ylabel('Amplitude')
axs[1].set_title('Reference signal')
axs[2].plot([0, I], [0, Q])
axs[2].set_xlabel('In-phase component')
axs[2].set_ylabel('Quadrature component')
axs[2].set_title('Demodulated signal')
plt.show()
```

```python
import numpy as np
from scipy.signal import hilbert
import matplotlib.pyplot as plt

# Define the parameters of the signal
omega_m = 2*np.pi*1000 # modulation frequency
gamma = 2*np.pi*100 # width

# Generate the rotation signal
t = np.linspace(0, 1, 10000)
phi_0 = 0.5
phi = phi_0*np.sin(omega_m*t)*np.exp(-t*gamma/2)

# Perform Hilbert transform to get in-phase and quadrature components
phi_hilbert = hilbert(phi)
phi_inphase = np.real(phi_hilbert)
phi_quadrature = np.imag(phi_hilbert)

# Plot the rotation signal and in-phase/quadrature components
fig, ax = plt.subplots(3, 1, figsize=(8, 6))

ax[0].plot(t, phi, label='Rotation signal')
ax[0].set_xlabel('Time (s)')
ax[0].set_ylabel('Rotation angle (rad)')
ax[0].legend()

ax[1].plot(t, phi_inphase, label='In-phase component')
ax[1].set_xlabel('Time (s)')
ax[1].set_ylabel('Amplitude')
ax[1].legend()

ax[2].plot(t, phi_quadrature, label='Quadrature component')
ax[2].set_xlabel('Time (s)')
ax[2].set_ylabel('Amplitude')
ax[2].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit

# Generate some sample data
x = np.linspace(0, 10, 1000)
y = 1/(1 + (x - 5)**2)

# Define the Lorentzian function to fit
def lorentz(x, a, b, c):
    return a/((x-b)**2 + (c/2)**2)

# Fit the Lorentzian to the data
popt, pcov = curve_fit(lorentz, x, y)

# Extract the parameters
a, b, c = popt
omega_m = b
gamma = c/2

# Plot the data and fit
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(x, y, label='Data')
ax.plot(x, lorentz(x, *popt), label='Fit')
ax.legend()
plt.show()

print(f'Central resonance modulation frequency: {omega_m:.2f}')
print(f'Width: {gamma:.2f}')
```
