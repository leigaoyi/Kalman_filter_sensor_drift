```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
record_num = 150

z_ref = np.loadtxt('../result/ZRef.txt')[:record_num]
kf_estimate_0 = np.loadtxt('../result/KF_y_delay0_145C_20Hz.txt')[:record_num]
kf_estimate_3 = np.loadtxt('../result/KF_y_delay3_145C_20Hz.txt')[:record_num]
kf_estimate_5 = np.loadtxt('../result/KF_y_delay5_145C_20Hz.txt')[:record_num]
kf_estimate_7 = np.loadtxt('../result/KF_y_delay7_145C_20Hz.txt')[:record_num]
kf_estimate_8 = np.loadtxt('../result/KF_y_delay8_145C_20Hz.txt')[:record_num]
kf_estimate_9 = np.loadtxt('../result/KF_y_delay9_145C_20Hz.txt')[:record_num]
kf_estimate_12 = np.loadtxt('../result/KF_y_delay12_145C_20Hz.txt')[:record_num]
kf_estimate_20 = np.loadtxt('../result/KF_y_delay20_145C_20Hz.txt')[:record_num]
time_range = np.array([1/1000 * i for i in range(len(z_ref))])
```

```python
# Set seaborn style
sns.set_style("white")
sns.set_palette("bone")

# Plot the yMeasure on the KF estimate
plt.figure()
plt.plot(time_range, z_ref, linestyle='-', linewidth=2, label='ideal response', color='red')
plt.plot(time_range, kf_estimate_0, linestyle='-', linewidth=2, label='KF Estimate without Delay', color='blue')
plt.plot(time_range, kf_estimate_3, linestyle='-', linewidth=2, label='KF Estimate with Delay', color='green')
# plt.plot(time_range, kf_estimate_0, 'o', markersize=5, linestyle='', label='Measurement', alpha=0.5, color='blue')
# plt.plot(time_range, kf_estimate_3, 'o', markersize=5, linestyle='', label='Measurement', alpha=0.5, color='green')
plt.title('yMeasure on KF Estimate')
plt.xlabel('Time (s)')
plt.ylabel('Magnetic Field (pT)')
plt.legend(loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set_style("white")
sns.set_palette("bone")

cropNum = 95

# Plot the yMeasure on the KF estimate
plt.figure()
plt.plot(1000*time_range[:cropNum], z_ref[:cropNum], linestyle='-', linewidth=2, label='Ref response', color='#DC5216')
plt.plot(1000*time_range[:cropNum], kf_estimate_0[:cropNum], linestyle='--', linewidth=2, label=r'$\tau=0$ ms', color='#0070BC', alpha=1.0)
plt.plot(1000*time_range[:cropNum], kf_estimate_8[:cropNum], linestyle='--', linewidth=2, label=r'$\tau=8$ ms', color='black', alpha=0.8)
#plt.plot(1000*time_range[:cropNum], kf_estimate_5[:cropNum], linestyle='--', linewidth=2, label=r'$\tau=5$ ms',color='green', alpha=0.8)
#plt.title('Time delay experiments')
plt.xlabel('Measurement Time (ms)')
plt.ylabel('Response (mV)')
plt.legend(loc='upper right', fontsize=12,frameon=False,bbox_to_anchor=(1.02, 1.0))
plt.savefig('TimeDelay.jpg', dpi=500)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set_style("white")
sns.set_palette("bone")

cropNum = 95

# Plot the yMeasure on the KF estimate
plt.figure()
ax = plt.gca()  # Get the current axes
plt.plot(1000*time_range[:cropNum], z_ref[:cropNum], linestyle='-', linewidth=2, label='Ideal response', color='#DC5216')
plt.plot(1000*time_range[:cropNum], kf_estimate_0[:cropNum], linestyle='--', linewidth=2, label=r'$\tau=0$ ms', color='#0070BC', alpha=1.0)
plt.plot(1000*time_range[:cropNum], kf_estimate_3[:cropNum], linestyle='--', linewidth=2, label=r'$\tau=3$ ms', color='black', alpha=0.8)

# Customize grid
ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.5, which='both')

#plt.title('Time delay experiments')
plt.xlabel('Measurement Time (ms)')
plt.ylabel('Response (mV)')
plt.legend(loc='upper right', fontsize=12, frameon=False, bbox_to_anchor=(1.02, 1.0))
#plt.savefig('TimeDelay.jpg', dpi=500)
plt.show()
```
