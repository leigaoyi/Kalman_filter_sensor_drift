```python
import matplotlib.pyplot as plt
import numpy as np

# Define the file name containing Chinese characters
file_name = u'../data/20230808/135C-小范围扫场-三角波-10Hz'

timeDelay = 4 # ms

# Read the file and extract the data columns. The first col1 is measure, col2 is input
data = np.loadtxt(file_name)
col1 = data[:, 0]
col2 = data[:, 1]

measure_data = col1
input_data = col2
```

```python
plt.plot(input_data[0:100])
```

```python
start_num = 13
# 20 Hz, 50 个点； 10Hz， 100个点

delta_Bx = input_data[800+start_num:850+start_num] * 1000 # mV
response = measure_data[805+start_num:855+start_num] * 1000
plt.plot(delta_Bx, response)
plt.grid()
plt.show()
# Linear fit
# fit = np.polyfit(delta_Bx, response, 1)
# slope = fit[0]
# print(slope)
# np.savetxt('../result/140C_large_sweep_x.txt', delta_Bx)
# np.savetxt('../result/140C_large_sweep_y.txt', response)
```

```python
# Linear fit
fit = np.polyfit(delta_Bx, response, 1)
slope = fit[0]

# Plotting
fig, ax = plt.subplots()
ax.plot(delta_Bx, response, 'o', label='Experimental Data', markersize=8)
ax.plot(delta_Bx, np.polyval(fit, delta_Bx), '-', label='Linear fit', linewidth=2)
ax.set_xlabel('$\delta B_x$ (mV)', fontsize=14)
ax.set_ylabel('Response (mV)', fontsize=14)
ax.legend(loc='best', fontsize=12)

# Customize grid appearance
ax.grid(True, which='both', linestyle='--', linewidth=0.5)
ax.minorticks_on()
ax.grid(which='minor', linestyle=':', linewidth=0.5)

# Remove right and top spines
#ax.spines['right'].set_visible(False)
#ax.spines['top'].set_visible(False)

ax.tick_params(axis='both', labelsize=12)

fig.tight_layout()

# Saving the figure
fig.savefig('../result/linear_fit.png', dpi=1000)
plt.show()

# Output slope
print('Slope of the linear fitting curve: {:.3f} mV/mV'.format(slope))
```
